```python
df_numeric = df2.copy()
df_numeric = df_numeric.apply(pd.to_numeric, errors='coerce')
df_numeric = df_numeric.dropna(axis=1, how='all')
corr = df_numeric.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Population Factors (2016-2022)")
plt.show()
```

```python
df["% Cases"]=df["Total Cases"]/df["Population"]*100
df["% Deaths"]=df["Total Deaths"]/df["Population"]*100
df["Deaths/Cases"]=df["Total Deaths"]/df["Total Cases"]
```

```python
myColumns = ['Population', 'Total Cases', 'Total Deaths', '% Cases',
       '% Deaths', 'Deaths/Cases']
```

```python
for column in myColumns:
  maxValues = df.nlargest(4,column)
  print("Top countries as per "+column)
  print(tabulate(maxValues, headers = 'keys', tablefmt = 'grid'))
```

```python
for column in myColumns:
  maxValues = df.nlargest(6,column)
  plt.figure(figsize=(25,6))
  plt.bar(maxValues['Country'], maxValues[column],color='lightblue')
  plt.show()
```

```python
for column in myColumns:
  maxValues = df.nsmallest(4,column)
  print("Top countries as per "+column)
  print(tabulate(maxValues, headers = 'keys', tablefmt = 'grid'))
```

```python
myColumns = ['% Cases','% Deaths', 'Deaths/Cases']
sns.set()
sns.pairplot(df[myColumns])
plt.show()
```

```python
#Understanding data based on countries
```

```python
df1 = df.set_index('Country')
df1
```

```python
corrData = df1.corr()
corrData
```

```python
meanValues = df1.mean()
meanValues
```

```python
varValues = df1.var()
varValues
```

```python
#Identifying Outliers
```

```python
for column in myColumns:
  outliers1 = df[df[column]>=meanValues[column]+2*math.sqrt(varValues[column])]
  outliers2 = df[df[column]<=meanValues[column]-2*math.sqrt(varValues[column])]
  outliers = pd.concat([outliers2,outliers1]);
  print("Outliers for "+column)
  print(outliers[["Country",column]])
```

```python
normalizedValues = pd.DataFrame()
for column in myColumns:
  mean = meanValues[column]
  std = math.sqrt(varValues[column])
  normalizedValues[column] = (df[column] - mean) / std
normalizedValues
```
